```python
# 绘制heatmap
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from mpmath import mp
import math

# 绘制32*32 heatmap
mp.dps = 1024 + 1
myPi = mp.pi
# 利用字符串分割，将小数点后数字取出
rest = str(myPi).split('.')
use_data = rest[-1]
print(len(use_data))

# 将数据转化为列表
A = list(use_data)
# 元素是字符串需要转化为int或float
A = [int(x) for x in A]
B = np.array([A]).reshape(32, 32)
print(B)

all_max = 9
all_min = 0
fig, axs = plt.subplots(1, 2, figsize =(64, 32))
plt.sca(axs[0])
ax = sns.heatmap(B, cmap = 'RdBu_r', vmax = all_max, vmin = all_min)
ax.set_aspect('equal')
plt.show()
```

```python
# pi后1000位各数字出现概率
from mpmath import mp
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

with st.sidebar:
    

    num_digits = st.slider('Number of decimal digits:',
                           min_value = 10000, 
                           max_value = 100000,
                           step = 10000)
    
mp.dps = num_digits + 2
pi_digits = mp.pi
# st.write(str(pi_digits))
pi_digits = str(pi_digits)[2:]
pi_digits_list = [int(x) for x in pi_digits]
# st.write(pi_digits_list)

pi_digits_array = np.array(pi_digits_list)

counts = np.bincount(pi_digits_array)

fig, ax = plt.subplots()

ax.barh(range(10), counts, align = 'center',
        edgecolor = [0.6, 0.6, 0.6])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlabel('Count')
ax.set_ylabel('Digit, 0~9')
plt.yticks(range(10))

st.pyplot(fig)

```

```python
# 判断整数奇偶性
num = float(input("Enter a number: "))
if num.is_integer():
    
    if (num % 2) == 0:
        print("{0} is even ".format(int(num)))
    else:
        print("{0} is odd ".format(int(num)))
else:
    print("{0} is not an integer".format(num))
```

```python
# 两种方法求矩阵加减
list1=[1, 2, 3]
list2=[4, 5, 6]
print([x + y for x, y in zip(list1, list2)])

print(list(map(lambda x,y:x+y, list1, list2)))
```
