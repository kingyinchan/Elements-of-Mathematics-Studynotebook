## 第三节 从代数式到函数

```python
# 函数的表达
from sympy.abc import x, y
expr = x**3 + 2*x**2 - x -2
print(expr.subs(x, 1)) 
```

利用.subs() 可以将x替换为其他变量，代数表达式。

```python
from sympy import cos
expr_cos_y = expr.subs(x,cos(y))
print(expr_cos_y)
```

```python
from sympy import symbols
x,y = symbols('x,y')

expr_1 = x + y
print(expr_1)

x1,x2 = symbols('x1,x2')
expr_2 = x1 + x2
print(expr_2)
```

```python
# 对表达式进行操作
from sympy import *
x,y,z = symbols('x,y,z')

# simplify mathematical expressions
expr_1 = sin(x)**2 - cos(x)**2
print(simplify(expr_1))

# expand plotnomial expressions 
expr_2 = (x + 1)**3
print(expand(expr_2))

# take a plotnomial and factors it into irreducible factors
expr_3 = x**3 + 2*x**2 - x - 2
print(factor(expr_3))

# collect common powers of a term in a expression
expr_collected = collect(expr_3 - x**2 - 2*x, x)
print(expr_collected)
```

```python
# 函数
from sympy.abc import x, y, z
expr = x**3 + 2*x**2 - x - 2

from sympy.utilities.lambdify import lambdify # lambdify是函数化工具（函数变量， 代数式）
f_x = lambdify(x, expr)

print(f_x(1))
```

## 第四节 杨辉三角

```python
# 使用火柴梗图表示杨辉三角
from sympy.abc import x, a
from sympy import Poly
import matplotlib.pyplot as plt
import numpy as np

for n in [4, 8, 12, 5, 9, 13, 36]:
    
    expr = (x + 1)**n
    
    expr_expand = expr.expand() # 展开
    expr_expand = Poly(expr_expand) # 将展开的写成多项式
    
    poly_coeffs = expr_expand.coeffs() # 获得多项式系数
    print(poly_coeffs)
    
    degrees = np.linspace(n,0,n + 1)
    
    fig, ax = plt.subplots() # 火柴梗图
    
    plt.stem(degrees, np.array(poly_coeffs, dtype=float))
    plt.xlim(0,n)
    plt.xticks(np.arange(0,n+1))
    y_max = max(poly_coeffs)
    y_max = float(y_max)
    plt.ylim(0,y_max)
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.invert_xaxis()
    plt.xlabel('Degree')
    plt.ylabel('Coefficients')
    
    plt.show()
```

## 第五节 排列组合

```python
# 无放回抽取组合实验
import itertools

letters = "ABC"

# find all combinations containing 2 letters

cmb = itertools.combinations(letters, 2)

for val in cmb:
    print(val)
```

```python
# 无放回排列实验
import itertools

letters = "ABC"

# Arranging 2 elements of 3

per = itertools.permutations(letters, 2)

for val in per:
    print(val)
```

```python
# 无放回全排列
import itertools

letters = "ABC"

# Arranging all 3 elements

per = itertools.permutations(letters, 3)

for val in per:
    print(val)
```
